```python
import pandas as pd
```

```python
train_cleaned = pd.read_csv('train_begin.csv')
```

```python
train_cleaned
```

```python
# 범주형 데이터를 원-핫 인코딩으로 변환
train_cleaned = pd.get_dummies(train_cleaned, columns=['GMES_ORIGIN_INSP_JUDGE_CODE Collect Result_AutoClave', 'GMES_ORIGIN_INSP_JUDGE_CODE Judge Value_AutoClave'])
```

```python
train_cleaned.describe(include='O')
```

```python
# 고유값이 1인 칼럼을 제거하는 코드
train_cleaned = train_cleaned.loc[:, train_cleaned.apply(lambda x: x.nunique() > 1)]
```

```python
train_cleaned
```

```python
train_cleaned.to_csv('train_second.csv', index=False)
```

```python
test_cleaned = pd.read_csv('test_begin.csv')
```

```python
test_cleaned
```

```python
# 범주형 데이터를 원-핫 인코딩으로 변환
test_cleaned = pd.get_dummies(test_cleaned, columns=['GMES_ORIGIN_INSP_JUDGE_CODE Collect Result_AutoClave', 'GMES_ORIGIN_INSP_JUDGE_CODE Judge Value_AutoClave'])
```

```python
test_cleaned.describe(include='O')
```

```python
# target 칼럼 제외하고 고유값이 1인 칼럼 드랍
target_column = 'target'
columns_to_drop = [col for col in test_cleaned.columns if col != target_column and test_cleaned[col].nunique() == 1]

# 드랍 수행
test_cleaned = test_cleaned.drop(columns=columns_to_drop)
```

```python
test_cleaned
```

```python
test_cleaned.to_csv('test_second.csv', index=False)
```
