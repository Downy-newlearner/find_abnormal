```python
import pandas as pd

train_second = pd.read_csv('C:/lg_capstone/find_abnormal/data/train_second.csv')
```

```python
# train_second에서 'Dam'으로 끝나는 컬럼명을 가진 컬럼들만 추출
Dam = train_second[train_second.columns[train_second.columns.str.endswith('Dam')][30:60]]
```

```python
Dam
```

```python
# Dam 데이터프레임에서 'Circle2, 3, 4'를 포함하는 컬럼명을 가진 컬럼들을 drop
Dam_temp = Dam.drop(Dam.columns[Dam.columns.str.contains('Circle2')], axis=1)
Dam_temp = Dam_temp.drop(Dam.columns[Dam.columns.str.contains('Circle3')], axis=1)
Dam_temp = Dam_temp.drop(Dam.columns[Dam.columns.str.contains('Circle4')], axis=1)
Dam_temp = Dam_temp.drop(Dam.columns[Dam.columns.str.contains('Line4')], axis=1)


Dam_shrinked = Dam_temp[Dam_temp.columns[3:]]
Dam_shrinked
```

```python
# Dam_shrinked 데이터프레임에서 'Stage1 Line1 Distance Speed Collect Result_Dam', 'Stage1 Line2 Distance Speed Collect Result_Dam', 'Stage1 Line3 Distance Speed Collect Result_Dam' 컬럼이 같으면 'Check_Stage1_Line' 컬럼에 true, 다르면 false를 저장
Dam_shrinked['Check_Stage1_Line'] = Dam_shrinked.apply(
    lambda row: row['Stage1 Line1 Distance Speed Collect Result_Dam'] == row['Stage1 Line2 Distance Speed Collect Result_Dam'] == row['Stage1 Line3 Distance Speed Collect Result_Dam'], 
    axis=1
)

Dam_shrinked['Check_Stage2_Line'] = Dam_shrinked.apply(
    lambda row: row['Stage2 Line1 Distance Speed Collect Result_Dam'] == row['Stage2 Line2 Distance Speed Collect Result_Dam'] == row['Stage2 Line3 Distance Speed Collect Result_Dam'], 
    axis=1
)

Dam_shrinked['Check_Stage3_Line'] = Dam_shrinked.apply(
    lambda row: row['Stage3 Line1 Distance Speed Collect Result_Dam'] == row['Stage3 Line2 Distance Speed Collect Result_Dam'] == row['Stage3 Line3 Distance Speed Collect Result_Dam'], 
    axis=1
)
```

```python
# 'Line1', 'Line3'을 포함한 컬럼 drop하여 Dam_shrinked_v2 데이터프레임 생성
Dam_shrinked_v2 = Dam_shrinked.drop(Dam_shrinked.columns[Dam_shrinked.columns.str.contains('Line1')], axis=1)

Dam_shrinked_v2 = Dam_shrinked_v2.drop(Dam_shrinked_v2.columns[Dam_shrinked_v2.columns.str.contains('Line3')], axis=1)
```

```python
Dam_shrinked_v2
```

```python
Dam_shrinked_v3 = Dam_shrinked_v2.copy()

# Dam_shrinked_v2 컬럼에서 'Stage1 Circle1 Distance Speed Collect Result_Dam', 'Stage1 Line2 Distance Speed Collect Result_Dam' 컬럼의 값이 행마다 같은지 확인
Dam_shrinked_v3['Check_Stage1'] = Dam_shrinked_v2.apply(
    lambda row: row['Stage1 Circle1 Distance Speed Collect Result_Dam'] == row['Stage1 Line2 Distance Speed Collect Result_Dam'], 
    axis=1
)

Dam_shrinked_v3['Check_Stage2'] = Dam_shrinked_v2.apply(
    lambda row: row['Stage2 Circle1 Distance Speed Collect Result_Dam'] == row['Stage2 Line2 Distance Speed Collect Result_Dam'], 
    axis=1
)

Dam_shrinked_v3['Check_Stage3'] = Dam_shrinked_v2.apply(
    lambda row: row['Stage3 Circle1 Distance Speed Collect Result_Dam'] == row['Stage3 Line2 Distance Speed Collect Result_Dam'], 
    axis=1
)
```

```python
# Dam_shrinked_v3에서 Check_Stage2 컬럼의 value count
Dam_shrinked_v3['Check_Stage2'].value_counts()
```

```python
# train_second에서 'Stage1 Circle'로 시작하는 컬럼명을 가진 컬럼들만 추출하여 데이터프레임 생성
train_second_Stage1_Circle = train_second[train_second.columns[train_second.columns.str.startswith('Stage1 Circle')]]

# train_second에서 'Stage2 Circle'로 시작하는 컬럼명을 가진 컬럼들만 추출하여 데이터프레임 생성
train_second_Stage2_Circle = train_second[train_second.columns[train_second.columns.str.startswith('Stage2 Circle')]]

# train_second에서 'Stage3 Circle'로 시작하는 컬럼명을 가진 컬럼들만 추출하여 데이터프레임 생성
train_second_Stage3_Circle = train_second[train_second.columns[train_second.columns.str.startswith('Stage3 Circle')]]

```

```python
# train_second_Stage1_Circle, train_second_Stage2_Circle, train_second_Stage3_Circle 각 데이터프레임들의 각 행에서 모든 컬럼 값이 일치하는지 확인

# 각 행마다 네 개의 열의 모든 데이터가 같은지 확인
train_second_Stage1_Circle['check'] = train_second_Stage1_Circle.apply(lambda x: x.nunique() == 1, axis=1)

train_second_Stage2_Circle['check'] = train_second_Stage2_Circle.apply(lambda x: x.nunique() == 1, axis=1)

train_second_Stage3_Circle['check'] = train_second_Stage3_Circle.apply(lambda x: x.nunique() == 1, axis=1)

```
