```python
import pandas as pd
import numpy as np
```

```python
train_second = pd.read_csv('C:/lg_capstone/find_abnormal/data/train_second.csv')
```

```python
# train_second의 "Model.Suffix"으로 시작하는 이름의 컬럼 모두 불러오기
Model_Suffix = train_second[train_second.columns[train_second.columns.str.startswith('Model.Suffix')]] #1
```

```python
Model_Suffix
```

```python
# train_second의 "Workorder"으로 시작하는 이름의 컬럼 모두 불러오기\
Workorder = train_second[train_second.columns[train_second.columns.str.startswith('Workorder')]]
```

```python
# train_second의 복사본 생성
train_second_copy = train_second.copy()

# train_second의 "Model.Suffix"으로 시작하는 이름의 컬럼이 모두 같은 값이므로 Model.Suffix라는 컬럼을 만들고 나머지는 삭제
train_second['Model.Suffix'] = Model_Suffix.iloc[:, 0]
train_second_copy = train_second.drop(Model_Suffix.columns, axis=1)

# train_second의 "Workorder"으로 시작하는 이름의 컬럼이 모두 같은 값이므로 Workorder라는 컬럼을 만들고 나머지는 삭제
train_second['Workorder'] = Workorder.iloc[:, 0]
train_second_copy = train_second.drop(Workorder.columns, axis=1)

# train_second의 Model.Suffix와 Workorder 컬럼만 포함한 데이터프레임 생성
train_second_copy = train_second_copy[['Model.Suffix', 'Workorder']]
train_second_copy
```

```python
# train_second의 Model.Suffix 컬럼의 값이 'AJX75334505'인 행만 추출하고 Workorder 컬럼의 unique 확인
AJX75334505 = train_second_copy[train_second_copy['Model.Suffix'] == 'AJX75334505']['Workorder'].unique()

# train_second_copy의 Model.Suffix 컬럼의 값이 'AJX75334501'인 행만 추출하고 Workorder 컬럼의 unique 확인
AJX75334501 = train_second_copy[train_second_copy['Model.Suffix'] == 'AJX75334501']['Workorder'].unique()

# train_second_copy의 Model.Suffix 컬럼의 값이 'AJX75334502'인 행만 추출하고 Workorder 컬럼의 unique 확인
AJX75334502 = train_second_copy[train_second_copy['Model.Suffix'] == 'AJX75334502']['Workorder'].unique()

# train_second_copy의 Model.Suffix 컬럼의 값이 'AJX75334507'인 행만 추출하고 Workorder 컬럼의 unique 확인
AJX75334507 = train_second_copy[train_second_copy['Model.Suffix'] == 'AJX75334507']['Workorder'].unique()

# train_second_copy의 Model.Suffix 컬럼의 값이 'AJX75334506'인 행만 추출하고 Workorder 컬럼의 unique 확인
AJX75334506 = train_second_copy[train_second_copy['Model.Suffix'] == 'AJX75334506']['Workorder'].unique()

# train_second_copy의 Model.Suffix 컬럼의 값이 'AJX75334508'인 행만 추출하고 Workorder 컬럼의 unique 확인
AJX75334508 = train_second_copy[train_second_copy['Model.Suffix'] == 'AJX75334508']['Workorder'].unique()

# train_second_copy의 Model.Suffix 컬럼의 값이 'AJX75334503'인 행만 추출하고 Workorder 컬럼의 unique 확인
AJX75334503 = train_second_copy[train_second_copy['Model.Suffix'] == 'AJX75334503']['Workorder'].unique()

# 위 7개의 넘파이 문자열 배열에서 겹치는 원소가 있는지 확인
np.intersect1d(np.intersect1d(np.intersect1d(np.intersect1d(np.intersect1d(np.intersect1d(AJX75334505, AJX75334501), AJX75334502), AJX75334507), AJX75334506), AJX75334508), AJX75334503)
# array([], dtype=object)
# 겹치는 원소가 없으므로 7개의 넘파이 문자열 배열은 모두 서로 다른 값이다. #2


```

```python
# train_second에서 AutoClave로 끝나는 컬럼만 불러오기
AutoClave = train_second[train_second.columns[train_second.columns.str.endswith('AutoClave')]]

# AutoClave에서 'Model.Suffix_AutoClave', 'Workorder_AutoClave', 'Chamber Temp. Judge Value_AutoClave' 열 drop
AutoClave = AutoClave.drop(['Model.Suffix_AutoClave', 'Workorder_AutoClave', 'Chamber Temp. Judge Value_AutoClave'], axis=1)

```

```python
# AutoClave 정규화
AutoClave = (AutoClave - AutoClave.mean()) / AutoClave.std()
AutoClave
```

```python
# 히트맵으로 상관계수 시각화
import seaborn as sns
import matplotlib.pyplot as plt

# AutoClave 데이터프레임에서 각 컬럼들의 서로에 대한 상관계수 구하기
AutoClave_corr = AutoClave.corr()

plt.figure(figsize=(10, 10))
sns.heatmap(AutoClave_corr, annot=True, cmap='coolwarm')
plt.show()
```

```python
# 컬럼 명에 'Pressure Unit Time'이 포함되는 컬럼만 불러오기
Pressure_Unit_Time = train_second[train_second.columns[train_second.columns.str.contains('Pressure Unit Time')]]
Pressure_Unit_Time
```

```python
# Pressure_Unit_Time 데이터프레임에서 '1st Pressure 1st Pressure Unit Time_AutoClave', '2nd Pressure Unit Time_AutoClave', '3rd Pressure Unit Time_AutoClave' 열의 값을 더하여 'Pressure Unit Time_AutoClave' 열 생성
Pressure_Unit_Time['Pressure Unit Time_AutoClave'] = Pressure_Unit_Time['1st Pressure 1st Pressure Unit Time_AutoClave'] + Pressure_Unit_Time['2nd Pressure Unit Time_AutoClave'] + Pressure_Unit_Time['3rd Pressure Unit Time_AutoClave']
Pressure_Unit_Time
```

```python
# Pressure_Unit_Time 데이터프레임에서 'Pressure Unit Time_AutoClave' 컬럼을 train_second 데이터프레임에 추가
train_second['Pressure Unit Time_AutoClave'] = Pressure_Unit_Time['Pressure Unit Time_AutoClave']

# train_second 데이터프레임에서 'target'값이 'Normal'이면 0, 'Abnormal'이면 1로 바꾸기
train_second['target'] = train_second['target'].replace({'Normal': 0, 'AbNormal': 1})

# train_second 데이터프레임에서 'Pressure Unit Time_AutoClave' 컬럼과 'target'컬럼의 상관관계 구하기
train_second[['Pressure Unit Time_AutoClave', 'target']].corr()
```

```python
# train_second 데이터프레임에서 'Chamber Temp. Judge Value_AutoClave' 컬럼과 'target' 컬럼만 불러오기
Chamber_Temp_Judge_Value_AutoClave = train_second[['Chamber Temp. Judge Value_AutoClave', 'target']]

# Chamber_Temp_Judge_Value_AutoClave 데이터프레임에서 'target' 값과 'Chamber Temp. Judge Value_AutoClave' 값이 같은 행만 불러오기\
# Chamber_Temp_Judge_Value_AutoClave[Chamber_Temp_Judge_Value_AutoClave['target'] == Chamber_Temp_Judge_Value_AutoClave['Chamber Temp. Judge Value_AutoClave']]
Chamber_Temp_Judge_Value_AutoClave
```
