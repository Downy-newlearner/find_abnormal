```python
from sklearn.preprocessing import LabelEncoder

# 문자열 데이터를 수치형으로 변환하는 함수
def encode_features(df):
    le = LabelEncoder()
    for column in df.columns:
        if df[column].dtype == 'object':
            df[column] = le.fit_transform(df[column])
    return df

# 데이터프레임에 타겟 변수를 추가
Dam_1['target'] = train_240_copy['target']
AC_1['target'] = train_240_copy['target']
Fill1_1['target'] = train_240_copy['target']
Fill2_1['target'] = train_240_copy['target']

Dam_2['target'] = train_1000_copy['target']
AC_2['target'] = train_1000_copy['target']
Fill1_2['target'] = train_1000_copy['target']
Fill2_2['target'] = train_1000_copy['target']

# 문자열 데이터 인코딩
Dam_1_encoded = encode_features(Dam_1)
AC_1_encoded = encode_features(AC_1)
Fill1_1_encoded = encode_features(Fill1_1)
Fill2_1_encoded = encode_features(Fill2_1)

Dam_2_encoded = encode_features(Dam_2)
AC_2_encoded = encode_features(AC_2)
Fill1_2_encoded = encode_features(Fill1_2)
Fill2_2_encoded = encode_features(Fill2_2)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from imblearn.pipeline import Pipeline as ImbPipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score
```

```python
# 데이터 불균형 문제
def sampling_error(X, y):
    smote = SMOTE(sampling_strategy='minority', random_state=123) # 소수 class over sampling
    under_sampler = RandomUnderSampler(sampling_strategy='majority', random_state=123) # 다수 class under sampling
    rf_model = RandomForestClassifier(random_state=123)
    pipeline = ImbPipeline(steps=[('o', smote), ('u', under_sampler), ('m', rf_model)]) # 순차적 학습 진행
    return pipeline

# target 변수 분리
def target_data(df):
    X = df.drop('target', axis=1)
    y = df['target']
    return X, y

X_Dam_1, y_Dam_1 = target_data(Dam_1)
X_AC_1, y_AC_1 = target_data(AC_1)
X_Fill1_1, y_Fill1_1 = target_data(Fill1_1)
X_Fill2_1, y_Fill2_1 = target_data(Fill2_1)

X_Dam_2, y_Dam_2 = target_data(Dam_2)
X_AC_2, y_AC_2 = target_data(AC_2)
X_Fill1_2, y_Fill1_2 = target_data(Fill1_2)
X_Fill2_2, y_Fill2_2 = target_data(Fill2_2)

model_Dam_1 = sampling_error(X_Dam_1, y_Dam_1)
model_AC_1 = sampling_error(X_AC_1, y_AC_1)
model_Fill1_1 = sampling_error(X_Fill1_1, y_Fill1_1)
model_Fill2_1 = sampling_error(X_Fill2_1, y_Fill2_1)

model_Dam_2 = sampling_error(X_Dam_2, y_Dam_2)
model_AC_2 = sampling_error(X_AC_2, y_AC_2)
model_Fill1_2 = sampling_error(X_Fill1_2, y_Fill1_2)
model_Fill2_2 = sampling_error(X_Fill2_2, y_Fill2_2)

model_Dam_1.fit(X_Dam_1, y_Dam_1)
model_AC_1.fit(X_AC_1, y_AC_1)
model_Fill1_1.fit(X_Fill1_1, y_Fill1_1)
model_Fill2_1.fit(X_Fill2_1, y_Fill2_1)

model_Dam_2.fit(X_Dam_2, y_Dam_2)
model_AC_2.fit(X_AC_2, y_AC_2)
model_Fill1_2.fit(X_Fill1_2, y_Fill1_2)
model_Fill2_2.fit(X_Fill2_2, y_Fill2_2)
```

```python
from sklearn.ensemble import VotingClassifier

# 앙상블 클래스
class CustomEnsemble:
    def __init__(self, models):
        self.models = models  # model을 받아 predict 메소드에 사용

    def predict(self, X):
        # 각 모델의 예측 결과를 저장
        predictions = []
        for model in self.models:
            try:
                pred = model.predict(X)
                predictions.append(pred)
            except Exception as e:
                print(f"Model {model} failed to predict: {e}")
        
        if not predictions:
            raise ValueError("No predictions available from the models.")

        # 가장 빈도수가 높은 예측 결과 선택
        return pd.DataFrame(predictions).mode(axis=0)[0]

# 모델 객체를 리스트로 수집 (각 모델을 리스트에 넣음)
models = [
    model_Dam_1, model_AC_1, model_Fill1_1, model_Fill2_1,
    model_Dam_2, model_AC_2, model_Fill1_2, model_Fill2_2
]

ensemble_model = CustomEnsemble(models=models)

# 테스트 데이터에 대한 예측
X_test = pd.concat([Dam_1_test, AC_1_test, Fill1_1_test, Fill2_1_test,
                    Dam_2_test, AC_2_test, Fill1_2_test, Fill2_2_test], axis=1)  # X_test는 테스트 데이터의 모든 특성을 포함

# 앙상블 모델 예측
y_pred_ensemble = ensemble_model.predict(X_test)

# 결과 출력 (예: 첫 번째 10개의 예측 결과 출력)
print(y_pred_ensemble.head(10))
```
